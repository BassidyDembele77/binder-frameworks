# Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?

```python
len(data['NTA'])
```

```python
d1=data[['NTA','Total Savings']].groupby(["NTA"]).mean()
d2=data[['NTA','Job created']].groupby(["NTA"]).sum()
```

```python
d1['Total created']=d2['Job created']
```

```python
d1
```

```python
d1.to_csv('day4.csv')
```

# scatter plot of jobs created versus average savings. Use both a standard and a logarithmic scale for the average savings.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.title('Job created versus Average savings')

plt.scatter(d1['Total Savings'], d1['Total created'])

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.title('Logarithmic Scale: Job created versus Average savings')

plt.scatter(np.log(d1['Total Savings']), d1['Total created'])

plt.show()
```

# histogram of the log of the average total savings.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.title('histogram of the log of the average total savings')
plt.hist(np.log(d1['Total Savings']), bins='auto')
```

# line plot of the total jobs created for each month.

```python
data['Month']=[data['Effective Date'][i][:2] for i in range(len(data))]
data_month=data[['Month','Job created']].groupby(["Month"]).sum()
```

```python
data_month['Job created']
```

```python
import matplotlib.pyplot as plt
plt.title('The  total jobs created for each month')
plt.xlabel('Month')
plt.ylabel('Total jobs created')

plt.plot(data_month.index, data_month["Job created"])

plt.show()
```
